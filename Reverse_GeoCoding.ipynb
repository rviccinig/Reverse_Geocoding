```python
import requests
import urllib.parse
import pandas as pd
import sys

#Opening File
Life_sciences_in_canada=pd.read_excel(open('Coordinates_Life_science_companies _Canada.xlsx','rb'),sheet_name='Sheet2', header=0,engine = 'openpyxl')

#Getting lists of lat/lon
lon=Life_sciences_in_canada['Value.longitude']
lat=Life_sciences_in_canada['Value.latitude']
```

```python
#Codigo Listo

Building_name_list=[]
Address_list=[]
City_list=[]
Province_list=[]
Postal_Code_list=[]
counter=0
    

for i,j in zip(lat,lon):
    url='https://nominatim.openstreetmap.org/reverse.php?lat='+str(i)+'&lon='+str(j)+'&zoom=18&format=jsonv2'
    response = requests.get(url).json()
    print(response["address"])
    # Normal Observation
    try:
        Building_name= response["address"]["building"]
        Adress=response["address"]["house_number"]+' '+ response["address"]["road"] +' '+response["address"]["neighbourhood"]
        City=response["address"]["city"] 
        Province=response["address"]["state"] 
        Postal_Code=response["address"]["postcode"]
    except:
        #If the building Name does not exist
        try:
            Building_name=" "
            Adress=response["address"]["house_number"]+' '+ response["address"]["road"] +' '+response["address"]["neighbourhood"]
            City=response["address"]["city"] 
            Province=response["address"]["state"] 
            Postal_Code=response["address"]["postcode"]
        except:
            try:
        #Building , Neighbordhood does no exist
                Building_name=" "
                Adress=response["address"]["house_number"]+' '+ response["address"]["road"]
                City=response["address"]["city"] 
                Province=response["address"]["state"] 
                Postal_Code=response["address"]["postcode"]
            except:
                try:
        #Building , Neighbordhood , and house number does no exist
                    Building_name=" "
                    Adress=response["address"]["road"]
                    City=response["address"]["city"] 
                    Province=response["address"]["state"] 
                    Postal_Code=response["address"]["postcode"]
                except:
                    try:
        #Building , Neighbordhood , and house number, city does no exist
                        Building_name=" "
                        Adress=response["address"]["road"]
                        City=response["address"]["state_district"] 
                        Province=response["address"]["state"] 
                        Postal_Code=response["address"]["postcode"]
                    except:
                        print(str(counter) + ' needs extra work')

    #Appending
    Building_name_list.append(Building_name)
    Address_list.append(Adress)
    City_list.append(City)
    Province_list.append(Province)
    Postal_Code_list.append(Postal_Code)
    counter+=1

```

```python
New_data=pd.DataFrame(list(zip(Building_name_list, Address_list, City_list,Province_list,Postal_Code_list)),columns=['Building_Name','Adress', 'City','Province','Postal_Code'])
```

```python
result = pd.concat([New_data, Life_sciences_in_canada], axis=1)
result.to_csv(r'Life_Sciences_Canada_2021.csv', index = False)
```
